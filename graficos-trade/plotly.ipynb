```python
import plotly

import plotly.offline as py
import plotly.graph_objs as go
import numpy as np

py.init_notebook_mode(connected=True)
```

```python
# Using graph_objects
fig = go.Figure([
    go.Scatter(x=df.date, y=df.close)
])
fig.show()
```

```python
import plotly.express as px

df['close_change'] = df.close.pct_change()

fig = px.bar(df, x=df[-100:].date, y=df[-100:].close_change)
fig.show()
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1)

fig.add_trace(
    go.Scatter(
        name="Price",
        mode="lines", x=df.date, y=df.close        
    ), row=1, col=1)

fig.add_trace(
    go.Scatter(
        name="Variation %",
        mode="lines", x=df.date, y=df.close_change
    ), row=2, col=1)

fig.update_layout(height=600, width=600, title_text="Stacked Subplots")
fig.show()
```

```python
fig = px.box(df.date, y=df.close_change)
fig.update_layout(height=600, width=600, title_text="Stacked Subplots")
fig.show()
```

```python
dft = df[0:100]
fig = go.Figure(data=[go.Candlestick(x=dft['date'],
                open=dft['open'],
                high=dft['high'],
                low=dft['low'],
                close=dft['close'])])

fig.update_layout(height=600, width=800, title_text="Stacked Subplots")
fig.show()
```
