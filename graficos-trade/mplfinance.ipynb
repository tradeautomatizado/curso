```python
df = pd.read_csv("../stocks/daily/VALE3.csv")
df.date = pd.to_datetime(df.date, format='%Y-%m-%d')
df = df.set_index(['date'])
```

```python
import mplfinance as mpf
```

```python
figsize = (15, 7)

df_100 = df[-100:]
mpf.plot(df_100, figsize=figsize)
```

```python
mpf.plot(df_100,type='candle',figsize=figsize)
```

```python
mpf.plot(df_100,type='line', figsize=figsize)
```

```python
mpf.plot(df[:-100],type='renko', figsize=figsize, renko_params=dict(brick_size=2))
```

```python
mpf.plot(df_100,type='candle',mav=(9,21,200),volume=True, figsize=(15, 10))
```
